```python
import os
import pandas as pd
import json
import requests
from pandas.io.json import json_normalize
from pprint import pprint

# API keys
from config import api_key
```

```python
# Save config information
url = "https://api.open.fec.gov/v1/elections/?api_key=[API]&sort=-total_receipts&sort_hide_null=false&office=president&election_full=true&sort_null_only=false&cycle=2016&sort_nulls_last=false&page=1"

# read data from the API 
response = requests.get(url).json()
# pprint(response)

# convert JSON to dataframe
candidate_finance = pd.json_normalize(response, "results")

# reorder columns
candidate_finance = candidate_finance[['candidate_election_year', 'candidate_id', 'candidate_name', 'candidate_pcc_name', 'party_full', 'total_receipts', 'total_disbursements', 'cash_on_hand_end_period', 'candidate_pcc_id', 'incumbent_challenge_full', 'committee_ids',  'coverage_end_date']]

# Format dataframe
candidate_finance.style.format({"total_receipts": '${0:,.2f}',
                                "total_disbursements": '${0:,.2f}',
                                "cash_on_hand_end_period": '${0:,.2f}',})
```
